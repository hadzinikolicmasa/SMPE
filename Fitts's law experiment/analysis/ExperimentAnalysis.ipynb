### Plotting the data with the linear regression and linear modelling for Experiment 1


```r
plot1 <- ggplot(meanMTdf_experiment1, aes(ID, MT)) +
  geom_point() +
  geom_smooth(method='lm', formula = y ~ x) +
  ggtitle("Experiment 1")

print(plot1)

model1 <- lm(MT~ID, data = meanMTdf_experiment1)
summary(model1)
```

### Conclusion

Based on the results of the experiments' analysis, we can conclude that there is a some positive relationship between ID and MT in computer experiments, both for the right and left hand. However, in the experiment with the phone, although there is a positive relationship between ID and MT, it is not  significant. These results indicate support for Fitts's law in cases involving computer use, while its application to phones appears less consistent. Further research may reveal specific factors contributing to these differences and enhance understanding of the application of Fitts's law in different technological contexts.

From this results, I can conclude that Fitts' law only takes into account the width and distance of the target, but not other factors that can affect performance and In my experiment we can see that factors such as device choice, concentration and hand can have an affect speed and accuracy.
